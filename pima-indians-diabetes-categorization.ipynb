```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
# for reproducibility of the random seed
numpy.random.seed(70)
```

```python
# dataset used http://archive.ics.uci.edu/ml/datasets/Pima+Indians+Diabetes
# keep dataset in the same directory of the program
# data file pima-indians-diabetes.data
dataset = numpy.loadtxt("pima-indians-diabetes.data", delimiter=",")

# split data into input (X) and output (Y) variables
X = dataset[:,:8]
Y = dataset[:,8]
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit model
model.fit(X,Y, epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X,Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
